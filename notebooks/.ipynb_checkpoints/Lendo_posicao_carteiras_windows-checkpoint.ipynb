```python
linhas_arquivo_ultimo_dia = []

with open(path_ultimo_dia, 'rt') as arquivo:
    for line in arquivo:
        line.strip()
        linhas_arquivo_ultimo_dia.append(line.replace('\x00','').strip().split('\t'))
```

```python
def posicao_acoes(linhas_arquivo):
    target_1 = ["Ação"]
    target_2 = 'Total'
    index_target_1 = 0
    index_target_2 = 0
    posicao = []
    
    for i in range(len(linhas_arquivo)):
        if linhas_arquivo[i] == target_1:
            index_target_1 = i
    
    for i in range(len(linhas_arquivo[index_target_1:])):
        if linhas_arquivo[index_target_1+i+2][0] == target_2:
            break
        elif linhas_arquivo[index_target_1+i+2] == ['']:
            pass
        else:
            posicao.append(linhas_arquivo[index_target_1+i+2])
            
    return posicao

```

```python
master_2021_03_17 = posicao_acoes(linhas_arquivo_ultimo_dia)

master_2021_03_17
```
