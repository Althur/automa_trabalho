```python
def xml_root(source):
    #tree = ET.parse(source)
    #root = tree.getroot()
    root = ET.fromstring(source)
    return root


def fetch_index_fundo(root):
    size = len(root)
    index_fundo = 0
    for i in range(size):
        if root[i].tag == 'fundo':
            print(i)
            index_fundo = i
            
    return index_fundo

def fetch_index_header(root_fundo):
    size = len(root_fundo)
    index_header = 0
    for i in range(size):
        if root_fundo[i].tag == 'header':
            print(i)
            index_header = i
    return index_header
    

def fetch_index_acao(root_fundo):
    size = len(root_fundo)
    #print(size)
    index_acao = []
    index_acao_start = 0
    index_acao_finish = 0
    
    for i in range(size):
        if root_fundo[i].tag == 'acoes':
            index_acao_start = i
            break
    index_acao.append(index_acao_start)
    
    for i in range(size):
        if root_fundo[i].tag == 'acoes':
            index_acao_finish = i
    
    index_acao.append(index_acao_finish)        
    
    
    return index_acao        
#    return index_acao_start, index_acao_finish

def fetch_codigo_ativo(root_ativo):
    size = len(root_ativo)
    index_codigo = 0
    
    for i in range(size):
        if root_ativo[i].tag == 'codativo':
            index_codigo = i
    return index_codigo


def fetch_index_quantidade(root_ativo):
    size = len(root_ativo)
    index_quantidade = 0 
    
    for i in range(size):
        if root_ativo[i].tag == 'qtdisponivel':
            index_quantidade = i
        elif root_ativo[i].tag == 'quantidade':
            index_quantidade = i
    return index_quantidade


def fetch_index_valor_fin(root_ativo):
    size  = len(root_ativo)
    index_valor_fin = 0
    
    for i in range(size):
        if root_ativo[i].tag == 'valorfindisp':
            index_valor_fin = i
    return index_valor_fin

def fetch_index_preco_unitario(root_ativo):
    size  = len(root_ativo)
    index_preco_unitario = 0
    
    for i in range(size):
        if root_ativo[i].tag == 'puposicao':
            index_preco_unitario = i
    return index_preco_unitario    

def fetch_index_classe_operacao(root_ativo):
    size  = len(root_ativo)
    index_classe_operacao = 0
    
    for i in range(size):
        if root_ativo[i].tag == 'classeoperacao':
            index_classe_operacao = i
    return index_classe_operacao



def fetch_index_vcto_btc(root_ativo):
    size  = len(root_ativo)
    index_vcto_btc = 0
    
    for i in range(size):
        if root_ativo[i].tag == 'dtvencalug':
            index_vcto_btc = i
    return index_vcto_btc


def fetch_index_taxa_btc(root_ativo):
    size  = len(root_ativo)
    index_taxa_btc = 0
    
    for i in range(size):
        if root_ativo[i].tag == 'txalug':
            index_taxa_btc = i
    return index_taxa_btc

```

```python
# pegar todas as infos, de acoes e jogar na lista
# filtrar oq for operação C, V e T
# ver se bat com a carteira da folhinha

def collect_infos(file):
    
    root = xml_root(file)
    index_fundo = fetch_index_fundo(root)
    index_acao = fetch_index_acao(root[index_fundo])
    index_ticker = fetch_codigo_ativo(root[index_fundo][index_acao[1]])
    index_quantidade = fetch_index_quantidade(root[index_fundo][index_acao[1]])
    index_classe_operacao = fetch_index_classe_operacao(root[index_fundo][index_acao[1]])
    
    info_list = []
    
    for i in range(index_acao[1]-index_acao[0]+1):
        lista_intermediaria = []
        ticker = root[index_fundo][index_acao[0]+i][index_ticker].text
        quantidade = int(root[index_fundo][index_acao[0]+i][index_quantidade].text)
        tipo_operação = root[index_fundo][index_acao[0]+i][index_classe_operacao].text
        
        lista_intermediaria.append(ticker)
        lista_intermediaria.append(quantidade)
        lista_intermediaria.append(tipo_operação)
        
        info_list.append(lista_intermediaria)
        
    for i in range(len(info_list)):
        print(info_list[i])
        
    return info_list

    
```
