```python
def posicao_acoes(linhas_arquivo):
    target_1 = ['Ação']
    target_2 = 'Total'
    index_target_1 = 0
    index_target_2 = 0
    posicao = []
    for i in range(len(linhas_arquivo)):
        if linhas_arquivo[i] == target_1:
            index_target_1 = i
        
    for i in range(len(linhas_arquivo[index_target_1:])):
        if linhas_arquivo[index_target_1+i+2][0] == target_2:
            break
        else:
            posicao.append(linhas_arquivo[index_target_1+i+2])
    
    return posicao

def posicao_emprestimo(linhas_arquivo):
    target_1 = ['Empréstimo de Ações']
    target_2 = 'Total'
    index_target_1 = 0
    index_target_2 = 0
    posicao = []
    for i in range(len(linhas_arquivo)):
        if linhas_arquivo[i] == target_1:
            index_target_1 = i
        
    for i in range(len(linhas_arquivo[index_target_1:])):
        if linhas_arquivo[index_target_1+i+2][0] == target_2:
            break
        else:
            posicao.append(linhas_arquivo[index_target_1+i+2])
    
    
    
    return posicao
```

```python
def gerar_linhas_arquivo(path):
    linhas_arquivo =[]
    linhas_arquivo_clean = []
    with open(path, "rt") as arquivo:
        for line in arquivo:
            line.strip()
            linhas_arquivo.append(line.replace('\x00','').strip().split('\t'))
            
    for i in range(len(linhas_arquivo)):
        if linhas_arquivo[i] == [""]:
            pass
        else:
            linhas_arquivo_clean.append(linhas_arquivo[i])
            
    return linhas_arquivo_clean
```
